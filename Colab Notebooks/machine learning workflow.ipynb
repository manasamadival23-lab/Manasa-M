```python
import pandas as pd
# Sample farm dataset
data = {
 "Rainfall_mm": [200, 180, 220, 210, 190],
 "Soil_Moisture": [30, 28, 35, 33, 29],
 "Yield_kg": [3200, 3000, 3500, 3400, 3100]
}
df = pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression
X = df[["Rainfall_mm", "Soil_Moisture"]]
y = df["Yield_kg"]
model = LinearRegression()
model.fit(X, y)
```

```python
predicted_yield = model.predict(X)
predicted_yield
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.lineplot(x=df.index, y='Yield_kg', data=df)
plt.title('Yield (kg) Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Yield (kg)')
plt.grid(True)
plt.show()
```
