```python
class Crop:
  def __init__(self,name,season,expected_yield):
    self.name=name
    self.season=season
    self.expected_yield=expected_yield

  def total_production(self,area):
    return self.expected_yield*area
```

```python
area=10
rice=Crop("Rice","Kharif",5.2)
wheat=Crop("Wheat","Rabi",4.1)

print("Total Production of Rice",rice.total_production(area))
print("Total Production of Wheat",wheat.total_production(area))
```

```python
rice=Crop("Rice","Kharif",5.2)
wheat=Crop("Wheat","Rabi",4.1)
corn=Crop("Corn","Rabi",3.5)
cotton=Crop("Cotton","Kharif",4.5)
sugarcane=Crop("Sugarcane","Rabi",2.5)
ragi=Crop("Ragi","Kharif",5.5)

crops=[rice,wheat,corn,cotton,sugarcane,ragi]
for crop in crops:
  production=crop.total_production(10)
  print(f"{crop.name}:{production}tons")
```

```python
class Livestock:
  def __init__(self, animal_type, age, weight):
    self.animal_type = animal_type
    self.age = age
    self.weight = weight

  def calculate_feed_requirement(self):
    return self.weight * 0.02
```

```python
cow = Livestock("Cow", 3, 500)
chicken = Livestock("Chicken", 0.5, 2)

print(f"A {cow.age}-year-old {cow.animal_type} weighing {cow.weight} kg requires {cow.calculate_feed_requirement():.2f} kg of feed per day.")
print(f"A {chicken.age}-year-old {chicken.animal_type} weighing {chicken.weight} kg requires {chicken.calculate_feed_requirement():.2f} kg of feed per day.")
```
