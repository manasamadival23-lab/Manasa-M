```python
for i in range(3):
    print(i)
```

```python
for i in range(2):
    for j in range(3):
        print(i, j)
```

```python
for i in range(3):
    for j in range(3):
        print("*", end=" ")
    print()
```

```python
rows = int(input("Enter rows: "))
cols = int(input("Enter columns: "))

for i in range(rows):
    for j in range(cols):
        print("#", end=" ")
    print()
```

```python
for i in range(3):
    j = 0
    while j < 2:
        print(i, j)
        j += 1
```

```python
for i in range(2):
    for j in range(3):
        print("i =", i, "j =", j)
        print(i, j)
```

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(i*j, end="\t")
    print()
```
