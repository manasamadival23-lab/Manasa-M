```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
yield_wheat = [200, 250, 300, 280, 320]

plt.plot(months, yield_wheat)
plt.xlabel("Month")
plt.ylabel("Yield (kg)")
plt.title("Wheat Yield Trend")
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
yield_wheat = [200, 250, 300, 280, 320]

plt.plot(months, yield_wheat)
plt.xlabel("Month")
plt.ylabel("Yield (kg)")
plt.title("Wheat Yield Trend")
plt.show()
```

```python
crops = ["Wheat", "Rice", "Maize", "Millet"]
yield_values = [3200, 2800, 3500, 3000]

plt.bar(crops, yield_values)
plt.xlabel("Crop")
plt.ylabel("Yield (kg)")
plt.title("Crop Yield Comparison")
plt.show()
```

```python
plt.bar(crops, yield_values,
        color=["green", "blue", "orange", "brown"])
plt.title("Crop Yield Comparison (Color Coded)")
plt.show()
```

```python
plt.plot(months,
         yield_wheat,
         marker="o")
plt.grid(True)
plt.show()
```

```python
plt.bar(crops, yield_values,
        color=["pink", "yellow", "red", "purple"])
plt.title("Crop Yield Comparison (Color Coded)")
plt.show()
```
