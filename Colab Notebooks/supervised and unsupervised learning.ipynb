```python
import numpy as np
from sklearn.linear_model import LinearRegression
# Training data: past seasons
rainfall = np.array([[200], [180], [220], [210]])
yield_kg = np.array([3200, 3000, 3500, 3400])
# Train supervised model
model = LinearRegression()
model.fit(rainfall, yield_kg)
# Predict yield for new season
model.predict([[205]])
```

```python
from sklearn.cluster import KMeans
import pandas as pd
# Soil data without labels
soil_data = pd.DataFrame({
"pH": [6.5, 6.8, 7.2, 7.5, 6.6],
"Moisture": [30, 35, 45, 50, 32]
})
# Group soil types automatically
kmeans = KMeans(n_clusters=2, random_state=0)
soil_data["Group"] = kmeans.fit_predict(soil_data)
soil_data
```

```python
from sklearn.cluster import KMeans
import pandas as pd
# Soil data without labels
soil_data = pd.DataFrame({
"pH": [6.5, 6.8, 7.2, 7.5, 6.6, 7.0, 6.9, 5.4],
"Moisture": [30, 35, 45, 50, 32, 44, 35, 56]
})
# Group soil types automatically
kmeans = KMeans(n_clusters=2, random_state=0)
soil_data["Group"] = kmeans.fit_predict(soil_data)
soil_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='pH', y='Moisture', hue='Group', data=soil_data, palette='viridis', s=100)
plt.title('Soil pH vs. Moisture by Cluster')
plt.xlabel('pH')
plt.ylabel('Moisture')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
