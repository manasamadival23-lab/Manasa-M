```python
import pandas as pd

# Crop health dataset
data = {
"Yellow_Leaves": [1, 1, 0, 0, 1, 0, 1, 0],
"Spots_Present": [1, 0, 0, 0, 1, 0, 0, 0],
"Disease": ["Fungal", "Nutrient", "Healthy",
"Healthy", "Fungal", "Healthy",
"Nutrient", "Healthy"]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.tree import DecisionTreeClassifier

X = df[["Yellow_Leaves", "Spots_Present"]] # Symptoms
y = df["Disease"] # Diagnosis

model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)
```

```python
# New plant observation
model.predict([[1, 1]])
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plot_tree(model,
feature_names=X.columns,
class_names=model.classes_,
filled=True)
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Add another data point to df
# Example: Adding a 'Healthy' case with no symptoms
new_data_point_2 = {"Yellow_Leaves": 0, "Spots_Present": 0, "Disease": "Healthy"}
new_row_df_2 = pd.DataFrame([new_data_point_2])
df = pd.concat([df, new_row_df_2], ignore_index=True)

# Re-define X and y with the updated DataFrame
X = df[["Yellow_Leaves", "Spots_Present"]] # Symptoms
y = df["Disease"] # Diagnosis

# Re-train the model with the updated data
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

print(f"Updated DataFrame (first 5 rows):\n{df.head()}\n")
print(f"Prediction for [0, 0] (no yellow leaves, no spots): {model.predict([[0, 0]])}")
```
