```python
import librosa

file_path = "/content/220727_211424.wav"  # Updated file path to an existing uploaded file

signal, sr = librosa.load(file_path, sr=None)

print("Sample Rate:", sr)
print("Signal Length:", len(signal))
```

```python
import numpy as np

mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=13)
mfcc_mean = np.mean(mfcc.T, axis=0)

print(mfcc_mean)
```

```python
features = []
labels = []

for file in os.listdir("/content/"):
    if file.endswith(".wav"):
        path = os.path.join("/content/", file)

        signal, sr = librosa.load(path, sr=None)
        mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=13)
        mfcc_mean = np.mean(mfcc.T, axis=0)

        features.append(mfcc_mean)

        # Labeling (IMPORTANT)
        if "queen" in file.lower():
            labels.append(1)
        else:
            labels.append(0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, random_state=42)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
def predict_audio(file_path):
    signal, sr = librosa.load(file_path, sr=None)
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc.T, axis=0)

    prediction = model.predict([mfcc_mean])

    if prediction[0] == 1:
        print("Queen Bee Present 🐝")
    else:
        print("Queen Bee Absent ❌")

predict_audio("/content/220727_211424.wav")
```
