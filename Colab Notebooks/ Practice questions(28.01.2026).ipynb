```python
rows = 5 # You can change the number of rows

for i in range(rows):
    if i == 0: # First row (apex)
        print(' ' * (rows - i - 1) + '*')
    elif i == rows - 1: # Last row (base)
        print('*' * (2 * rows - 1))
    else: # Intermediate rows
        spaces_before = ' ' * (rows - i - 1)
        spaces_between = ' ' * (2 * i - 1)
        print(spaces_before + '*' + spaces_between + '*')
```

```python
rows = 5 # You can change the number of rows

for i in range(1, rows + 1):
    if i == 1:
        print('*')
    elif i == rows:
        print('*' * rows)
    else:
        print('*' + ' ' * (i - 2) + '*')
```
