```python
#Arithmetic Operators Use Case:

print("====== FARMER INPUT BASED PROFIT ANALYSIS ======")

seed_cost = float(input("Enter seed cost (INR): "))
fertilizer_cost = float(input("Enter fertilizer cost (INR): "))
pesticide_cost = float(input("Enter pesticide cost (INR): "))
labor_cost = float(input("Enter labor cost (INR): "))
irrigation_cost = float(input("Enter irrigation cost (INR): "))

total_input_cost = seed_cost + fertilizer_cost + pesticide_cost + labor_cost + irrigation_cost
print("\nTotal Input Cost (INR):", total_input_cost)

total_yield_kg = float(input("\nEnter total crop yield (kg): "))
price_per_kg = float(input("Enter selling price per kg (INR): "))

total_revenue = total_yield_kg * price_per_kg
print("Total Revenue (INR):", total_revenue)

profit = total_revenue - total_input_cost
print("Net Profit / Loss (INR):", profit)

profit_per_kg = profit / total_yield_kg
print("Profit per Kg (INR):", profit_per_kg)

total_seed_kg = float(input("\nEnter total seed available (kg): "))
seed_bag_capacity = float(input("Enter seed bag capacity (kg): "))

full_seed_bags = total_seed_kg // seed_bag_capacity
remaining_seed = total_seed_kg % seed_bag_capacity

print("Full Seed Bags Used:", full_seed_bags)
print("Remaining Seed (kg):", remaining_seed)

initial_crop_units = float(input("\nEnter initial crop units: "))
growth_rate = float(input("Enter seasonal growth rate (e.g., 1.10 for 10%): "))
seasons = int(input("Enter number of seasons: "))

expected_crop = initial_crop_units * (growth_rate ** seasons)
print("Expected Crop After Seasons:", expected_crop)

weather_loss = float(input("\nEnter loss due to weather (INR, enter positive value): "))
weather_loss = -weather_loss

adjusted_profit = profit + weather_loss
print("Adjusted Profit After Weather Impact (INR):", adjusted_profit)

positive_view = +adjusted_profit
negative_view = -adjusted_profit

print("Positive Profit View:", positive_view)
print("Negative Profit View:", negative_view)

print("\n====== END OF FARMER ANALYSIS ======")
```

```python
import matplotlib.pyplot as plt

# Retrieve values from the notebook state
initial_crop_units = 1000.0  # From your notebook state
growth_rate = 1.15         # From your notebook state
seasons_count = 2          # From your notebook state (number of seasons to simulate)
profit_per_kg = 18.5       # From your notebook state

season_numbers = list(range(seasons_count + 1)) # Including season 0 (initial)
profit_over_seasons = []

print("Calculating hypothetical profit for each season:")
for s in season_numbers:
    # Calculate expected crop for the current season 's'
    # Assuming initial_crop_units is for season 0, and subsequent seasons grow.
    expected_crop_at_season = initial_crop_units * (growth_rate ** s)
    # Calculate hypothetical profit for this season
    profit_at_season = expected_crop_at_season * profit_per_kg
    profit_over_seasons.append(profit_at_season)
    print(f"Season {s}: Expected Crop = {expected_crop_at_season:.2f} kg, Hypothetical Profit = {profit_at_season:.2f} INR")

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(season_numbers, profit_over_seasons, marker='o', linestyle='-', color='blue')
plt.title('Hypothetical Profit Trend Over Seasons')
plt.xlabel('Season Number')
plt.ylabel('Hypothetical Profit (INR)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(season_numbers) # Ensure all season numbers are shown on x-axis
plt.tight_layout()
plt.show()
```

```python
#Arithmetic Operators
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b) 		# Divides and returns the integer part only
print("Modulus:", a % b)
print("Exponentiation:", a ** b)		# Raises the first number to the power of the second (ab)
print("Unary Plus:", +a)
print("Unary Minus:", -a)
```

```python
#Membership Operators

crops = ["rice", "wheat", "maize"]

print("rice in crops :", "rice" in crops)
print("barley in crops :", "barley" in crops)
print("barley not in crops :", "barley" not in crops)
```

```python
#Logical Operators
rainfall = True
soil_moisture = False

print("AND:", rainfall and soil_moisture)
print("OR :", rainfall or soil_moisture)
print("NOT:", not rainfall)

```

```python
#Identity Operators

a = [1, 2, 3]
b = a
c = [1, 2, 3]

print("a is b :", a is b)		 #TRUE
print("a equals c :", a == c)     #TRUE
print("a is c :", a is c)		 #FALSE
print("a is not c :", a is not c) #TRUE
```

```python
#Comparison (Relational) Operators

x = 20
y = 15

print("x == y :", x == y)
print("x != y :", x != y)
print("x > y  :", x > y)
print("x < y  :", x < y)
print("x >= y :", x >= y)
print("x <= y :", x <= y)
```

```python
#Bitwise Operators
a = 10
b = 6

print("Bitwise AND (&):", a & b)
print("Bitwise OR (|):", a | b)
print("Bitwise XOR (^):", a ^ b)
print("Bitwise NOT (~a):", ~a)
print("Left Shift (a << 1):", a << 1)
print("Right Shift (a >> 1):", a >> 1)
```

```python
#Assignment Operators

a = 10
print("Initial value:", a)

a += 5				#a=a+5
print("After += :", a)		#a=10+5=15

a -= 3				#a=a-3
print("After -= :", a)		#a=15-3=12

a *= 2				#a=a*2
print("After *= :", a)		#a=12*2=24

a /= 4				#a=a/4
print("After /= :", a)		#a=24/4=6

a //= 2				#a=a/2 (floor value)
print("After //= :", a)		#a=6/2=3

a %= 3				#a=a/3 (remainder)
print("After %= :", a)		#a=3/3 (remainder=0)

a **= 2				#a=a2
print("After **= :", a)		#a=02 = 0
```
