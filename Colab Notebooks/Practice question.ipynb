```python
# Question 7: Yield Prediction Based on Conditions

# --- Input Collection ---
try:
    rainfall = float(input("Enter rainfall in mm: "))
    temperature = float(input("Enter temperature in Celsius: "))
    soil_quality = float(input("Enter soil quality (80-100 points scale): "))
except ValueError:
    print("Invalid input. Please enter numeric values for rainfall, temperature, and soil quality.")
    # Exit or return to prevent further execution with invalid input
    # In a notebook, 'exit()' might not prevent buffered prints, so we'll structure code to prevent execution
    exit()

# --- Input Validation ---
# Only proceed with calculations if soil quality is valid
if not (80 <= soil_quality <= 100):
    print("Error: Soil quality must be between 80 and 100 points.")
else:
    # --- Initialization ---
    base_yield = 40.0 # quintals/hectare
    current_yield = base_yield

    # --- Rainfall Adjustment ---
    rainfall_adjustment_percentage = 0.0
    if rainfall > 600:
        above_600 = rainfall - 600
        rainfall_adjustment_percentage = (above_600 // 10) * 2 # +2% per 10mm
    elif rainfall < 400:
        below_400 = 400 - rainfall
        rainfall_adjustment_percentage = -(below_400 // 10) * 2 # -2% per 10mm

    current_yield *= (1 + rainfall_adjustment_percentage / 100)
    print(f"\nYield after Rainfall adjustment ({rainfall_adjustment_percentage:+2.0f}%): {current_yield:.2f} quintals/hectare")

    # --- Temperature Adjustment ---
    temperature_adjustment_percentage = 0.0
    if temperature < 25:
        degrees_off = 25 - temperature
        temperature_adjustment_percentage = -degrees_off * 1 # -1% per degree
    elif temperature > 30:
        degrees_off = temperature - 30
        temperature_adjustment_percentage = -degrees_off * 1 # -1% per degree

    current_yield *= (1 + temperature_adjustment_percentage / 100)
    print(f"Yield after Temperature adjustment ({temperature_adjustment_percentage:+2.0f}%): {current_yield:.2f} quintals/hectare")

    # --- Soil Quality Adjustment ---
    # Soil quality is proportional to final yield, so it's a direct multiplier
    soil_quality_multiplier = soil_quality / 100
    current_yield *= soil_quality_multiplier
    print(f"Yield after Soil Quality adjustment ({soil_quality:.0f}%): {current_yield:.2f} quintals/hectare")

    # --- Final Output ---
    print(f"\nFinal Predicted Yield: {current_yield:.2f} quintals/hectare")
```

```python
# Question 8: Fertilizer Selection Based on Soil Test Results

# --- Input Collection ---
try:
    n_level = float(input("Enter Nitrogen (N) level in mg/kg: "))
    p_level = float(input("Enter Phosphorus (P) level in mg/kg: "))
    k_level = float(input("Enter Potassium (K) level in mg/kg: "))
except ValueError:
    print("Invalid input. Please enter numeric values for N, P, and K levels.")
    exit()

# --- Define Thresholds ---
# Assuming a threshold to classify as 'Low' or 'High' for NPK
# For simplicity, we'll use 50 mg/kg as the common threshold.
# Below this is 'Low', equal or above is 'High'.
THRESHOLD = 50.0

# --- Determine Fertilizer Recommendation ---
recommendation = ""

# Helper booleans for clearer conditions
is_n_low = n_level < THRESHOLD
is_p_low = p_level < THRESHOLD
is_k_low = k_level < THRESHOLD

if is_n_low and is_p_low and is_k_low:
    recommendation = "Use NPK 10-26-26"
elif is_n_low and not is_p_low and not is_k_low:
    recommendation = "Use Urea only"
elif not is_n_low and is_p_low and is_k_low:
    recommendation = "Use DAP (18-46-0)"
elif not is_n_low and not is_p_low and is_k_low:
    recommendation = "Use Muriate of Potash"
else:
    # This covers cases like High N, High P, High K, or other combinations not specified
    # The problem implies 'All balanced' as the fallback.
    recommendation = "No fertilizer needed (Soil levels appear balanced or not needing specific intervention)"

# --- Output Results ---
print(f"\nSoil Test Results: N={n_level}mg/kg, P={p_level}mg/kg, K={k_level}mg/kg")
print(f"Recommendation: {recommendation}")
print(f"(Based on assumed threshold of {THRESHOLD} mg/kg for Low/High classification)")
```

```python
# Question 9: Water Management Decision

# --- Input Collection ---
try:
    rainfall_current_month = float(input("Enter rainfall received this month (in mm): "))
    rainfall_predicted_next_month = float(input("Enter predicted rainfall next month (in mm): "))
    available_water_in_tank = float(input("Enter available water in tank (in liters): "))
    crop_water_requirement = float(input("Enter crop water requirement for next month (in liters): "))
except ValueError:
    print("Invalid input. Please enter numeric values for all inputs.")
    exit()

# --- Define Constants and Thresholds (Adjustable) ---
TANK_CAPACITY = 2000.0 # liters (example value, adjust as per farm's tank capacity)
LOW_TANK_THRESHOLD_PERCENT = 25
MEDIUM_TANK_THRESHOLD_PERCENT = 50
HIGH_TANK_THRESHOLD_PERCENT = 75

RAINFALL_SLIGHT_DEFICIT_THRESHOLD = 50 # liters/mm

# --- Calculations ---
total_expected_rainfall_water = (rainfall_current_month + rainfall_predicted_next_month) * 10 # Assuming 1mm rainfall over 1 sq meter area is 1 liter
rainfall_balance = total_expected_rainfall_water - crop_water_requirement

tank_fill_percentage = (available_water_in_tank / TANK_CAPACITY) * 100

# --- Decision Logic ---
management_strategy = ""

print(f"\n--- Water Management Summary ---")
print(f"Current Rainfall: {rainfall_current_month} mm")
print(f"Predicted Rainfall: {rainfall_predicted_next_month} mm")
print(f"Available Tank Water: {available_water_in_tank} liters ({tank_fill_percentage:.1f}% full)")
print(f"Crop Water Requirement: {crop_water_requirement} liters")
print(f"Expected Rainfall Contribution: {total_expected_rainfall_water} liters")

if rainfall_balance >= 0: # Rainfall meets or exceeds requirement
    if tank_fill_percentage >= HIGH_TANK_THRESHOLD_PERCENT:
        management_strategy = "Excellent conditions. Rely on abundant rainfall. Consider storing excess water or using it for expansion if feasible."
    else:
        management_strategy = "Good conditions. Rainfall is sufficient. Conserve tank water for potential dry spells or unexpected needs."
elif rainfall_balance < 0 and rainfall_balance >= -RAINFALL_SLIGHT_DEFICIT_THRESHOLD: # Slight rainfall deficit
    if tank_fill_percentage >= MEDIUM_TANK_THRESHOLD_PERCENT:
        management_strategy = "Slight rainfall deficit. Supplement crop water requirement from tank as needed. Monitor conditions."
    else:
        management_strategy = "Slight rainfall deficit. Tank levels are moderate to low. Prioritize conservation. Supplement from tank carefully or explore minor alternative sources."
else: # Significant rainfall deficit (rainfall_balance < -RAINFALL_SLIGHT_DEFICIT_THRESHOLD)
    if tank_fill_percentage >= HIGH_TANK_THRESHOLD_PERCENT:
        management_strategy = "Significant rainfall deficit. Utilize tank water to meet crop needs. Monitor predicted rainfall carefully for any changes."
    elif tank_fill_percentage >= MEDIUM_TANK_THRESHOLD_PERCENT:
        management_strategy = "Significant rainfall deficit. Tank water is moderate. Use tank water judiciously, prioritize critical crops, and prepare for potential water scarcity."
    elif tank_fill_percentage >= LOW_TANK_THRESHOLD_PERCENT:
        management_strategy = "High deficit. Tank water is low. Implement strict conservation measures. Strongly consider seeking external water sources or reducing water-intensive crops."
    else:
        management_strategy = "Severe water shortage. Tank is critically low. Immediate and strict conservation measures are essential. Secure external water sources urgently to prevent crop failure."

# --- Output Results ---
print(f"\nWater Management Strategy: {management_strategy}")
```
