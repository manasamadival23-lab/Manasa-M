```python
number = int(input("Enter a number:"))

if number % 2 == 0:
    print(f"The number {number} is Even.")
else:
    print(f"The number {number} is Odd.")
```

```python
# Define base fares
FULL_FARE = 300.00
HALF_FARE = 150.00
DISCOUNT_FARE = 200.00

try:
    age = int(input("Please enter the passenger's age: "))

    if age < 0:
        print("Age cannot be negative. Please enter a valid age.")
    elif age < 5:
        fare = 0.00
        category = "Child"
    elif age <= 12:
        fare = HALF_FARE
        category = "Kid"
    elif age <= 17:
        fare = DISCOUNT_FARE
        category = "Teen"
    elif age <= 64:
        fare = FULL_FARE
        category = "Adult"
    else:
        fare = DISCOUNT_FARE
        category = "Senior"

    if age >= 0:
        print(f"For a passenger of {age} years old ({category}), the bus fare is: Rs{fare:.2f}")

except ValueError:
    print("Invalid input. Please enter a whole number for age.")
```

```python
# Example of a 'while' loop
# This loop counts from 1 up to 3
count = 1
print("Counting up to 3:")
while count <= 3:
    print(count)
    count += 1
```
