```python
#2.IRRIGATION FREQUENCY OF CROPS
crop=input("enter the crop name:")
soil_moisture=int(input("enter soil moisture in %:"))
temp=input("enter temperature condition:")
print(f"Crop:{crop}, Soil moisture:{soil_moisture}%, Temperature condition:{temp}")
if crop == "Rice" and soil_moisture < 60 and temp == "hot" or soil_moisture < 50 and temp == "moderate" or temp == "cool":
    print(f"Decision: IRRIGATE NOW \nReason:soil moisture ({soil_moisture})% is below threshold (50%) for moderate. \nEstimated water needed:50mm.")
elif crop == "Wheat" and soil_moisture < 40 and temp == "hot" or soil_moisture < 30 and temp == "moderate" or temp == "cool":
    print(f"Decision: IRRIGATE NOW \nReason:soil moisture ({soil_moisture})% is below threshold (30%) for moderate. \nEstimated water needed:40mm")
elif crop == "Sugarcane" and soil_moisture < 70 and temp == "hot" or soil_moisture < 60 and temp == "moderate" or temp == "cool":
    print(f"Decision: IRRIGATE NOW \nReason:soil moisture ({soil_moisture})% is below threshold (60%) for moderate. \nEstimated water needed:50mm")
else:
    print("Decision: DO NOT IRRIGATE")
```

```python
#3.PEST SEVERITY LEVEL
p=int(input("enter the number of pests:"))
d=int(input("enter the crop damage percentage in %:"))
t=int(input("enter the temp in °C:"))
h=int(input("enter the humidity in %:"))
print(f"Pest per plant: {p}, Damage: {d}%, temp: {t}°C, Humidity: {h}%")
if p<5 and d<10:
  print("Severity Level: LOW \nRecommendation: No action required")
elif p>=5 and p<=15 and d>=10 and d<=25:
  print("Severity Level: MEDIUM \nRecommendation: Monitor daily. \nConsider targeted pesticide spray if severity increases. \nCost of Intervention: Rs500/ha")
else:
  print("Severity Level: HIGH \nRecommendation: Immediate action required. \nConsider targeted pesticide, spray immediatly. \nCost of Intervention: Rs1000/ha")
```

```python
base_yield = 40

rain = int(input("Enter rainfall (mm): "))
temp = int(input("Enter temperature (°C): "))
soil = int(input("Enter soil quality: "))

# Rainfall factor
if rain > 600:
    rain_factor = 1 + 0.02 * ((rain - 600) / 10)
elif rain < 400:
    rain_factor = 1 - 0.02 * ((400 - rain) / 10)
else:
    rain_factor = 1

# Temperature factor
if temp < 25:
    temp_factor = 1 - 0.01 * (25 - temp)
elif temp > 30:
    temp_factor = 1 - 0.01 * (temp - 30)
else:
    temp_factor = 1

# Soil quality factor
if soil < 80:
    soil_factor = 1 - 0.01 * (80 - soil)
else:
    soil_factor = 1

final_yield = base_yield + rain_factor + temp_factor + soil_factor

print("Predicted Crop Yield =", round(final_yield, 2), "quintals/hectare")
```
