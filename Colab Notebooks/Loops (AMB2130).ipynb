```python
for i in range(2):
  for j in range (3):
    print("i=",i,"j=",j)
    print(i,j)
```

```python
for i in range (1,6):
  for j in range(1,6):
    print(i*j,end="\t")
    print()
```

```python
for i in range(3):
  for j in range(3):
    print("*",end="")
    print()
```

```python
for i in range(1, 6):
  for j in range(i):
    print("*", end="")
  print()
```

```python
for i in range(5, 0, -1):
  for j in range(i):
    print("*", end="")
  print()
```

```python
rows = 5
for i in range(1, rows + 1):
  print(" " * (rows - i), end="")
  print("* " * i)
```

```python
num = 7
for i in range(1, 11):
  print(f"{num} x {i} = {num * i}")
```
